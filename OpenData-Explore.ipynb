```python
label_columns = [','.join(measures[int(column.replace('F', ''))]) for column in df.columns ]
len(label_columns)
```

```python
labeled_df = pd.DataFrame(df)
labeled_df.columns = label_columns
labeled_df.head(5)
```

```python
import pickle

image_labels = pickle.load(open('data/image_labels.pkl', 'rb'))
len(image_labels), image_labels['A1.jpg']
```

```python
labels = list()
for value in labeled_df.index.values:
    
    source_labels = image_labels[value.replace('/', '') + ".jpg"]
    if not source_labels or len(source_labels) == 0:
        continue
        
    labels.append({
        'id': value, 
        'race.label': source_labels['race.label'], 
        'race.confidence': source_labels['race.confidence'], 
        'gender.label': source_labels['gender.label'], 
        'gender.confidence': source_labels['gender.confidence']}
    )
    
consolidated_df = pd.concat([labeled_df, pd.DataFrame(labels).set_index('id')], axis=1)
consolidated_df.head(5)
```

```python
confidence_cutoff = .80
confident_consolidated_df = consolidated_df[(consolidated_df['race.confidence'] >= confidence_cutoff) & (consolidated_df['gender.confidence'] >= confidence_cutoff)]
consolidated_df.shape, confident_consolidated_df.shape
```

```python
confident_consolidated_df.to_csv('data/opendata-with-labels.csv')
```
